### **Block 1**

```python
#load following dataset using given code
import pandas as pd

from sklearn.datasets import load_diabetes

diabetes = load_diabetes()

X = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)

y = diabetes.target

X.info()
```

## **Block 2**

```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## **Block 3**

```python
#create a function that: 
#3a. takes X and y as in the input
#3b. Split the X and y into train and test using train_test_split from sklearn
#3c. fit a linear regression model to the training X and training y data
#3d. get the predicted y data for X test from the model
#3e. Output (at least) the model fit (i.e.., R^2 score = coefficient)

def model_data(X,y):
    Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,random_state=1)

    model = LinearRegression(fit_intercept=True)
    model.fit(Xtrain,ytrain)

    print('Training R^2',model.score(Xtrain,ytrain))
    print('Testing R^2',model.score(Xtest,ytest))
    print('y = ',np.round(model.coef_[0], 3),'*x + ',np.round(model.intercept_,3))

    y_model = model.predict(Xtest)

    sns.scatterplot(x = Xtest[Xtest.columns[0]],y = ytest)
    sns.lineplot(x = Xtest[Xtest.columns[0]],y = y_model,color='r')
```

## **Block 4**

```python
#show that this function works (i.e. try it with some of the data)
model_data(X[['age']],y)
```

## **Block 5**

```python
#use the function to find the single best feature (ie. column) in X that best predicts y
model_data(X[['age']],y)
print("\n")
model_data(X[['sex']],y)
print("\n")
model_data(X[['bmi']],y)
print("\n")
model_data(X[['bp']],y)
print("\n")
model_data(X[['s1']],y)
print("\n")
model_data(X[['s2']],y)
print("\n")
model_data(X[['s3']],y)
print("\n")
model_data(X[['s4']],y)
print("\n")
model_data(X[['s5']],y)
print("\n")
model_data(X[['s6']],y)
print("\n")

print("the single best feature is s5 (highest testing R^2)")
print(model_data(X[['s5']],y))
```

## **Block 6**

```python
#6a. make a figure that has two subplots using object oriented programming
#6b. On the left, put the training data as a scatterplot (x,y).
#6c. On the right, the testing data as a scatterplot (x,y).
#6d. Include the best fitting model line in both plots and include the R^2 value in both plots for the different fits.

def model_data2(X,y):
  Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,random_state=1)

  model = LinearRegression(fit_intercept=True)
  model.fit(Xtrain,ytrain)
  
  print('Training R^2',model.score(Xtrain,ytrain))
  print('Testing R^2',model.score(Xtest,ytest))
  print('y = ',model.coef_[0],'*x + ',model.intercept_)
  
  y_model = model.predict(Xtest)

  fig,axes = plt.subplots(Xtrain.shape[1],1,figsize=[8,12])
  fig,axes2 = plt.subplots(Xtest.shape[1],1,figsize=[8,12])

  if Xtrain.shape[1] == 1:
    sns.scatterplot(x = Xtest[Xtest.columns[0]],y = ytest,ax=axes)
    sns.lineplot(x = Xtest[Xtest.columns[0]],y = y_model,ax=axes)
  elif Xtest.shape[1] == 1:
    sns.scatterplot(x = Xtrain[Xtrain.columns[0]],y = ytest,ax=axes)
    sns.lineplot(x = Xtrain[Xtrain.columns[0]],y = y_model,ax=axes)
    return
  
  
model_data2(X[['s5']],y)
```
