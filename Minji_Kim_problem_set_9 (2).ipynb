## **BLOCK 1**

```python
#import relevant libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#import data using seaborn
data = sns.load_dataset('geyser')

#check the dataset
data
```

## **BLOCK 2**

```python
#define x and y
x = data[['duration','waiting']]
y = [1 if k == 'long' else 2 for k in data['kind']]

#create k-fold cross validation code
def kfold(N):
  #create empty list of chunks
  x_chunks = []
  y_chunks = []
  chunk_size = x.shape[0] // N
  #first for loop creates set of chunks and fills the list up
  for i in range(N):
    start = i * chunk_size
    if i != N-1:
      x_chunks.append(x[start:start+chunk_size])
      y_chunks.append(y[start:start+chunk_size])
    else:
      x_chunks.append(x[start:])
      y_chunks.append(y[start:])

  scores = []
  model = LinearRegression()
  
  for i in range(N):
     # build the training set by picking N-1 chunks
     #the chunk that gets left out is chunk number i, and goes from 0 to N-1
     x_train = pd.concat([x_chunks[c] for c in range(N) if c != i])
     # remaining chunk gets selected as the test set
     x_test = x_chunks[i]
     y_train = []
     y_test = y_chunks[i]
     for c in range(N):
        if c!= i:
          y_train.extend(y_chunks[c])
    #do linear regression model fitting
     model.fit(x_train, y_train)
     scores.append(model.score(x_test, y_test))

  return scores

#kfold cross validation for # of folds being 10
kf = kfold(10)
display(kf)
```

## **BLOCK 3**

```python
#print average and standard deviation for the different folds
narr = np.array(kf)
print('Mean for kfold(10):')
print(np.mean(narr))
print('STD for kfold(10):')
print(np.std(narr))
print('\n')

narr2 = np.array(kfold(3))
print('Mean for kfold(3):')
print(np.mean(narr2))
print('STD for kfold(3):')
print(np.std(narr2))
print('\n')

narr3 = np.array(kfold(5))
print('Mean for kfold(5):')
print(np.mean(narr3))
print('STD for kfold(5):')
print(np.std(narr3))
print('\n')

narr4 = np.array(kfold(20))
print('Mean for kfold(20):')
print(np.mean(narr4))
print('STD for kfold(20):')
print(np.std(narr4))
```

## **BLOCK 4**

The standard deviation is relatively low, and everything is close to the mean. Every score lies within approximately one standard deviation unit of the mean score, which means regardless of what section of the data is chosen as our training set and testing set, we will get approximately the same result. 
Additionally, the mean is relatively close to 1. This means that there is a high predictiveness between duration of an eruption in minutes and the duration in minuutes until the next eruption, and it is a good predictor.
