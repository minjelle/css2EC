## **Block 2**

```python
#2a. load data and create a dataframe called "data"
print('------------ 2a ------------')
files.upload()
data = pd.read_csv('problem_set_1_data.csv')
print(data)

#2b. #rename all columns
print('------------ 2b ------------')
data.columns = ['quiz_total', 'attended', 'final_grade', 'final_exam', 'name']
print(data)

#2c. #print "head" of df
print('------------ 2c ------------')
print(data.head())
print('\n')

#2d. #print "tail" of df
print('------------ 2d ------------')
print(data.tail())
print('\n')

#2e. #print summary of df
print('------------ 2e ------------')
print(data.describe())
```

The averages from the dataframe can be analyzed in the "mean" row. In 'quiz_total', the scores were low, average to around a 69%. However, for final grade, the scores averaged out to be around a 91%. This means that people did considerably well on their final exam(which was most likely out of 50pts), and people scored on average 39pts. Overall, students did fairly well in the class, even if quiz scores were low. 



## **Block 3**

```python
#subset the data by whether someone attended class
#3a. 
print('------------ 3a ------------')
print(data[data['attended']==1])
print('\n')
print(data[data['attended']==0])

#find averages for those who attended class vs. did not attend
#3b. 
print('------------ 3b ------------')
print(data[data['attended']==1].describe())
print('\n')
print(data[data['attended']==0].describe())
```

For those who attended class:
*   quiz_total avg: 72.50
*   final_grade avg: 95.07
*   final_exam avg: 40.64

For those who did not attend class:
*   quiz_total avg: 68.33
*   final_grade avg: 91.09
*   final_exam avg: 39.44


According to the data shown, there does not seem to be a statistically significant difference between in grades of those who attend or do not attend. It was not significant enough to change their letter grade to a lower letter. Although those who attended class scored a few points higher, those who did not attend were able to maintain their grade just fine. Additionally, it also shows that the quizzes were most likely hard, regardless of lecture attendance. It seems apparent that those who did not attend class were able to receive outside help of some sort, and pass this class.
