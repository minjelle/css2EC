## **Block 2**

```python
#web scraping the data
#web scrap using pd.read_html
webdata = pd.read_html('https://en.wikipedia.org/wiki/MonsterVerse')

webdata = webdata[3]
webdata
```

## **Block 3**

```python
#clean up the data frame pt.1
#3a. make the index the diff characters instead of 0, 1, 2, 3
#print('------------ 3a ------------')
webdata.index = webdata['Character']

#3b. change the column names to just be the names of the movies; change the index name to also just be “character”
webdata.columns = ['Character', 'Godzilla','Kong: Skull Island','Godzilla: King of the Monsters', 'Godzilla vs. Kong']

#3c. remove the row that has all duplicates of 'Humans'
webdata.drop(webdata.index[5])
webdata.drop(columns='Character', inplace = True)
```
